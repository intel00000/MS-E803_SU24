```python
# load GridSearch_MLPRegressor_output.xlsx
import pandas as pd
df = pd.read_excel('GridSearch_MLPRegressor_output.xlsx', engine='openpyxl')
```

```python
# show the columns
print(df.columns)
```

```python
# get the "Unnamed: 0" column
print(df['Unnamed: 0'])
```

```python
# grep the hidden_layer_sizes in the "Unnamed: 0" column in format {'hidden_layer_sizes': XXX}
import re
pattern = re.compile(r"{'hidden_layer_sizes': (\d+)}")
hidden_layer_sizes = df['Unnamed: 0'].apply(lambda x: pattern.search(x).group(1))
hidden_layer_sizes
```

```python
# now grep the mean_test_score
mean_test_score = df['mean_test_score'].__neg__()
```

```python
# plot the the mean_test_score against hidden_layer_sizes
import matplotlib.pyplot as plt
plt.plot(hidden_layer_sizes, mean_test_score)
plt.xlabel('hidden_layer_sizes')
plt.ylabel('mean_test_score')
plt.show()
```
